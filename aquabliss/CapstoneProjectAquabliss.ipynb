```python
def is_student_on_course(s,c):
    a = df1[df1.StudentId==s]
    b = a[a.ClassId==c]
    if b.shape[0] == 0:
        return False
    else: 
        return True
    
```

```python
def class_details(class_id):
    a = df_class[df_class.Id==class_id]
    if a.shape[0] == 0:
        raise KeyError
    
    return(a.StartDate.iloc[0], a.LessonTime.iloc[0], a.TeacherId.iloc[0])

   
       
```

```python
def start_lesson_teacher(StartDate, LessonTime, TeacherId):
    a = df_class[(df_class.StartDate==StartDate) & (df_class.LessonTime == LessonTime) & (df_class.TeacherId == TeacherId)]
    if a.shape[0] == 0:
        raise KeyError
    return a.Id.iloc[0]
                 
    
```

```python
#Loading dataset and cleaning data
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import statsmodels.api as sm
from sklearn import datasets, linear_model
%matplotlib inline
conn = sqlite3.connect('E:/MGMCapstone Project - Aquabliss/everywhere.db')
cur = conn.cursor()
#Since we are predicting whether squad based learning is anything different to being in a learn-to-swim class, 
#we shall select the table classtypes as a better representation of the study models. 
cur.execute("select * from ClassLevels;")
df= pd.read_sql_query("select * from ClassLevels; ", conn)
df.head()
```

```python
#in order to understand which class is "better" than the other, we shall qeury the columns of priority and activeness
df= pd.read_sql_query("select * from ClassLevelSkills; ", conn)
df[['ClassLevel', 'Skill', 'Priority','Active']]
df.head()
```

```python
#Plotting selected columns within ClassTypes to gain a comprative outlook of the selected study variables
df.plot(x='ClassLevel', y='Priority', kind='bar')
plt.title("Class levels that recieve highest priority from instructors")
```

```python
fig, ax = plt.subplots(1, 1)
df.plot(x='ClassLevel', y='Priority', label="Best Time Allocation", ax=ax)
plt.title("Class levels that recieve adequate training time from instructors")
```

```python
w=df.Priority
z=df.Active
z=sm.add_constant(z)
z.head()
```

```python
#We shall use Linear Regression to predict squad based learning is anything different to being in a learn-to-swim class
est=sm.OLS(w, z)
est=est.fit()
est.summary()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
df= pd.read_sql_query("select * from ClassLevelSkills; ", conn)
df.head() 
X = df['Priority'].values[:,np.newaxis]
y = df['Active'].values
model2 = LinearRegression()
model2.fit(X, y)
plt.title('Squad vs. Learn to swim class')
plt.xlabel('Squard Predicted')
plt.ylabel('Learn to Swim Actual')
plt.scatter(X, y,color='g')
plt.plot(X, model2.predict(X),color='k')
plt.show()
```
